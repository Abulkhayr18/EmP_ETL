```python
from dotenv import load_dotenv
import os

load_dotenv()

DATABRICKS_HOST  = os.getenv("DATABRICKS_HOST")
DATABRICKS_TOKEN = os.getenv("DATABRICKS_TOKEN")
HTTP_PATH        = os.getenv("DATABRICKS_HTTP_PATH")

# Verify they loaded
print("Host loaded:", DATABRICKS_HOST is not None)
print("Token loaded:", DATABRICKS_TOKEN is not None)
print("HTTP Path loaded:", HTTP_PATH is not None)
```

```python
from databricks import sql

# Connect to Databricks
with sql.connect(
    server_hostname=DATABRICKS_HOST,
    http_path=HTTP_PATH,
    access_token=DATABRICKS_TOKEN
) as conn:
    cursor = conn.cursor()

    #  Load employees table 
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS employees (
            EMPLOYEE_ID BIGINT,
            NAME STRING,
            DEPARTMENT STRING,
            AGE BIGINT,
            SALARY BIGINT,
            YEARS_OF_EXPERIENCE BIGINT,
            CITY STRING,
            SalaryWithTax DOUBLE
        )
    """)

    emp_rows = [tuple(row) for row in employees.itertuples(index=False, name=None)]
    cursor.executemany("INSERT INTO employees VALUES (?,?,?,?,?,?,?,?)", emp_rows)
    print(f" employees loaded  {len(emp_rows)} rows")

    #  Load sales table 
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS sales (
            OrderID STRING,
            Product STRING,
            Quantity BIGINT,
            Price BIGINT,
            Region STRING,
            SalesRep STRING
        )
    """)

    sales_rows = [tuple(row) for row in sales.itertuples(index=False, name=None)]
    cursor.executemany("INSERT INTO sales VALUES (?,?,?,?,?,?)", sales_rows)
    print(f" sales loaded  {len(sales_rows)} rows")

    cursor.close()

print(" All done!")
```
